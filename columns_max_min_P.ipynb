```python
df = pd.read_excel("4B^MG^M15_R32_SS_R8_Column Forces Extraction_1.xlsx")
df.head()
```

```python
df["P"] = df["P"].abs()
df.head()
```

```python
df["Key"] = df["Story"] + df["Column"]
df
```

```python
maxx = df.groupby("Key").P.idxmax()
maxx
```

```python
minn = df.groupby("Key").P.idxmin()
minn
```

```python
df_max = df.loc[maxx]
df_max["max/min"] = "max"
df_min = df.loc[minn]
df_min["max/min"] = "min"
```

```python
df_out = pd.concat([df_max,df_min])
df_out
```

```python
df_out.to_excel("out.xlsx")
```
